```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date

plt.style.use('ggplot')
%matplotlib inline
```

```python
ALLEN_AI_RES_PATH = "../res/allenai/metadata.csv"
DIMENSIONS_RES_PATH = "../res/dimensions/export.csv"
```

```python
df_allen = pd.read_csv(ALLEN_AI_RES_PATH)

print(len(df_allen.index))
df_allen.head(5)
```

```python
df_dim = pd.read_csv(DIMENSIONS_RES_PATH)

print(len(df_dim.index))
df_dim.head(5)
```

```python
df_dim.columns.tolist()
```

```python
print('DOI')
print(df_allen.fillna(-1).groupby('doi').filter(lambda x: len(x) > 1).doi.value_counts())
print('----')
print(df_dim.fillna(-1).groupby('DOI').filter(lambda x: len(x) > 1)['DOI'].value_counts())
```

```python
print('Title + Authors (identity)')
print(len(df_allen[df_allen.duplicated(['title', 'authors'])].index), "potential duplicates within allen")
df_allen[df_allen.duplicated(['title', 'authors'], keep=False)].sort_values('title').head(8)
```

```python
print('Title + Authors (identity)')
print(len(df_dim[df_dim.duplicated(['Title', 'Authors'])].index), "potential duplicates within dimensions")
df_dim[df_dim.duplicated(['Title', 'Authors'], keep=False)].sort_values('Title').head(8)
```

```python
print('DOI in Allen, but not in Dimensions:', len(df_allen[~df_allen['doi'].isin(df_dim['DOI'])].index))
print('DOI In Dimensions, but not in Allen:', len(df_dim[~df_dim['DOI'].isin(df_allen['doi'])].index))
```

```python
print('Title in Allen, but not in Dimensions:', len(df_allen[~df_allen['title'].isin(df_dim['Title'])].index))
print('Title In Dimensions, but not in Allen:', len(df_dim[~df_dim['Title'].isin(df_allen['title'])].index))
```

```python
df_doi_overlap = df_dim[df_dim['DOI'].isin(df_allen['doi'])]
print(len(df_doi_overlap.index), 'of', len(df_dim.index), 'DOIs from dimensions also in Allen')
```
